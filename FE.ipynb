```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('Inventory_SupplyChain_Dataset.csv')
df.head()
```

```python
# Financial KPI: Inventory Turnover Ratio
df['Inventory_Turnover'] = df['Cost of Goods Sold (COGS)'] / df['Average Inventory']

print(df['Inventory_Turnover'].describe())
```

```python
# Operational KPI: Warehouse Utilization Rate
df['Warehouse_Utilization'] = df['Inventory Level'] / df['Warehouse Capacity']
df['Warehouse_Utilization'].describe()
```

```python
# Efficiency KPI: Logistics Cost per Unit
df['Cost_Per_Unit_Sold'] = df['Transportation Cost'] / df['Units Sold']
df['Cost_Per_Unit_Sold'].describe()
```

```python
# Logical Feature: Stock Status Flag
df['Stock_Status'] = np.where(df['Warehouse_Utilization'] > 0.8, 'High Risk', 
                     np.where(df['Warehouse_Utilization'] < 0.2, 'Critical Low', 
                     'Optimal'))

df['Stock_Status'].describe()
```
